```python
# Assuming list_of_a contains elements with 'text' and 'href' attributes (from BeautifulSoup)
for a in list_of_a:
    # Open the CSV file in append mode ('a+')
    with open("url.csv", "a+") as f:
        # Extract the text and link
        extracted_text = a.text
        extracted_link = "https://en.wikipedia.org" + a["href"]
        
        # Write the extracted text and link, separated by a comma (for CSV format)
        f.write(f"{extracted_text}, {extracted_link}\n")

f.close()
```
