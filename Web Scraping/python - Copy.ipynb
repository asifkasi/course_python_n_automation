## Requests

```python
import requests
```

```python
def fun_hit_url(url):
    r = requests.get(url)
    if r.status_code == 200:
        from bs4 import BeautifulSoup

        soup = BeautifulSoup(r.text, 'html.parser')
        
        return soup
    else:    
        return False
```
